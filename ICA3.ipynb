```python
#let's read the lego sets HB file
filename = "/content/drive/MyDrive/DATA/lego_setsHB.csv"
counter=0
price = {}
difficulty = {}
with open(filename, 'r') as f1:
    next(f1) #will help skip the first line in which we have the headers
    for line in f1:
      tokens = line.split(',')
      price[tokens[4]] = tokens[0]
      difficulty[tokens[4]] = tokens[5]
      counter+=1
      if counter==1000:
        break
for key, value in price.items():
    print(key, value, difficulty[key])
```
